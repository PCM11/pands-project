```python
# Sepal length histogram
sns.histplot(x = "sepal_length", data = iris, hue="species", multiple="stack")
plt.title("Sepal Length", size=10, color="red")
#plt.savefig("Sepal length")
```

```python
# Sepal width histogram
sns.histplot(x = "sepal_width", data = iris, hue="species", multiple="stack")
plt.title("Sepal Width", size=10, color="red")
plt.savefig("Sepal width")
```

```python
#Petal length vs Sepal length plot
sns.scatterplot(y = "petal_length", x = "sepal_length", data = iris, hue = "species")
                
#Title.
plt.title(("Petal length vs Sepal length plot"), color = "red")
plt.savefig("Petal length vs Sepal length plot.png")
```

```python
#Petal width vs Sepal width plot
sns.scatterplot(y = "petal_width", x = "sepal_width", data = iris, hue = ("species"))
                
#Title.
plt.title(("Petal width vs Sepal width plot"), color = "red")
plt.savefig("Petal width vs Sepal width plot.png")
```

```python
# Sepal length vs Sepal width plot

sns.scatterplot( y = "sepal_length", x = "sepal_width", data = iris, hue = "species")
                
#Title.
plt.title(("Sepal length vs Sepal width plot"), color = "red")

plt.savefig("Sepal length vs Sepal width plot.png")
```

```python
#Petal length vs Petal width plot
sns.scatterplot(y = "petal_length", x = "petal_width", data = iris, hue = ("species"))
                
#Title.
plt.title(("Petal length vs Petal width plot"), color = "red")
plt.savefig("Petal length vs Petal width plot.png")
```

```python
# Pairplot
sns.pairplot(iris, hue = "species", diag_kind="hist")
```

```python
sns.histplot(x = "petal_length", data = iris, hue="species", multiple = "stack")
plt.title("Petal Length", size=10, color="red")

```

```python

# Petal width histogram
sns.histplot(x = "petal_width", data = iris, hue="species", multiple="stack")
plt.title("Petal Width", size=10, color="red")
plt.savefig("Petal width")
```

```python
#iris = iris.drop(["species"], axis = 1) #We must first drop the species column.
iris.corr().round (2) # round up the values to two digits for easy reading
```

```python
# Let's plot the correlation matrix map.
matrix = iris.corr().round (2)
sns.heatmap(matrix, annot=True, vmax=1, vmin=-1, center=0, cmap='vlag')
plt.tittle("Correlation Matrix")
```

```python
sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
```
