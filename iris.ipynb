```python
# Data frames.
import pandas as pd

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns

#Numerical arrays
import numpy as np
```

# Load Data
***

```python
# Fetch data from URL.
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
print(iris) # displays the whole dataset

#print(iris.head()) # displays the 5 first rows from the dataframe

#print(iris.describe()) # diplays stats about the data

#print(iris.info()) # displays basic information about datatype

#print(iris['species'].value_counts()) # displays number of samples in each class
```

```python
# Petal Variable
plen = iris['petal_length'] # to get the petal length
plen = plen.to_numpy() # to get the numpy array
#print(plen)

pwidth = iris['petal_width']
pwidth =pwidth.to_numpy() # petal width to array
#print(pwidth)
```

```python
# Sepal length vs Sepal width plot

sns.scatterplot( y = "sepal_length", x = "sepal_width", data = iris, hue = "species")
                
#Title.
plt.title(("Sepal length vs Sepal width plot"), color = "red")

plt.savefig("Sepal length vs Sepal width plot.png")
```

```python
#Petal length vs Petal width plot
sns.scatterplot(y = "petal_length", x = "petal_width", data = iris, hue = ("species"))
                
#Title.
plt.title(("Petal length vs Petal width plot"), color = "red")
plt.savefig(" Petal length vs Petal width plot.png")
```

```python
#Sepal length vs Petal length plot
sns.scatterplot(y = "petal_length", x = "sepal_length", data = iris, hue = ("species"))
                
#Title.
plt.title(("Sepal length vs Petal length vs Sepal length plot"), color = "red")
plt.savefig("Sepal length vs Petal width plot.png")
```

```python
#Petal width vs Sepal width plot
sns.scatterplot(x = "petal_width", y = "sepal_width", data = iris, hue = ("species"))
                
#Title.
plt.title(("Petal width vs Sepal width plot"), color = "red")
```
